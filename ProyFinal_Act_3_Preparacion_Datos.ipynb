3. Filtrado de valores an칩malos:

Aplicaremos un filtro b치sico para eliminar posibles valores an칩malos en las columnas de Height y Weight.

```python
# Filtrado de valores an칩malos
data_filtered = data[(data['Height'] >= 1.2) & (data['Height'] <= 2.5)]
data_filtered = data_filtered[(data_filtered['Weight'] >= 30) & (data_filtered['Weight'] <= 200)]
```

```python
data_filtered.head()
```

```python
# Verificar la distribuci칩n de clases en el dataset filtrado
clases_originales = data_filtered['NObeyesdad'].value_counts()
print(f"Distribuci칩n original de clases en NObeyesdad:\n{clases_originales}")
```

4. Normalizaci칩n

Normalizaremos las columnas num칠ricas Age, Height, y Weight.

```python
from sklearn.preprocessing import StandardScaler

# Normalizar columnas num칠ricas
scaler = StandardScaler()
data_filtered[['Age', 'Height', 'Weight']] = scaler.fit_transform(data_filtered[['Age', 'Height', 'Weight']])
```

5. Codificaci칩n de Variables Categ칩ricas

Convertimos las variables categ칩ricas utilizando one-hot encoding.

```python
# Codificaci칩n de variables categ칩ricas
data_encoded = pd.get_dummies(data_filtered, drop_first=True)
```

```python
# Revisar la codificaci칩n one-hot
data_encoded = pd.get_dummies(data, columns=['NObeyesdad'])
print(f"Columnas generadas por la codificaci칩n one-hot:\n{data_encoded.columns}")
```

```python
# Supongamos que estas son las columnas categ칩ricas que necesitas codificar
columnas_categoricas = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

# Aplicar la codificaci칩n one-hot
data_encoded = pd.get_dummies(data, columns=columnas_categoricas)


```
